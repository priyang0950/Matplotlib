3. Write a Python program to draw line charts of the financial data of Alphabet Inc. between October 3, 2016 to October 7, 2016. 

```python
import datetime as dt

dates = [
    dt.date(2016, 10, 3),
    dt.date(2016, 10, 4),
    dt.date(2016, 10, 5),
    dt.date(2016, 10, 6),
    dt.date(2016, 10, 7)
]

prices = [772.56, 776.43, 776.47, 776.86, 775.08]

plt.plot(dates, prices, marker='o')

plt.xlabel("Date")
plt.ylabel("Closing Price ")
plt.title("Alphabet Inc. Stock Prices (Oct 3–7, 2016)")
plt.tight_layout()
plt.grid(True)
plt.show()
```

19. Write a Python program to create bar plot of scores by group and gender. Use multiple X values on the same chart for men and women. 
Sample Data: 
Means (men) = (22, 30, 35, 35, 26) 
Means (women) = (25, 32, 30, 35, 29) 

```python
men = (22, 30, 35, 35, 26)
women = (25, 32, 30, 35, 29)

x = np.arange(5)
width = 0.35

plt.bar(x, men, width, label='Men',color='green')
plt.bar(x + width, women, width, label='Women',color='red')

plt.xlabel('Group')
plt.ylabel('Scores')
plt.title('scores gender')
plt.xticks(x, ['G1', 'G2', 'G3', 'G4', 'G5'])
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample salary data generate kar rahe hain
np.random.seed(42)
salaries = np.random.normal(loc=65000, scale=20000, size=3000)

# Salary range limit
salaries = np.clip(salaries, 20000, 110000)

# Mean calculate karna
mean_salary = np.mean(salaries)

# Plot histogram
plt.figure(figsize=(10,6))
plt.hist(salaries, bins=20, color='#39b5d4', edgecolor='black')

# Mean line
plt.axvline(mean_salary, color='black', linestyle='dashed', linewidth=2)
plt.text(mean_salary + 2000, 220, 'Mean', fontsize=12)

# Labels and title
plt.title("Salary Distribution", fontsize=16)
plt.xlabel("Salary")
plt.ylabel("Frequency")

# X-axis ticks formatting
plt.xticks(np.arange(20000, 115000, 5000), rotation=45)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Reproducible data
np.random.seed(10)

# Right-skewed salary distribution generate karna
salaries = np.random.lognormal(mean=10.4, sigma=0.4, size=2000)

# Salary range limit
salaries = np.clip(salaries, 20000, 110000)

# Mean & Median calculate
mean_salary = np.mean(salaries)
median_salary = np.median(salaries)

# Plot
plt.figure(figsize=(10,6))
plt.hist(salaries, bins=20, color='#39b5d4', edgecolor='black')

# Mean line
plt.axvline(mean_salary, color='black', linestyle='dashed', linewidth=2)
plt.text(mean_salary + 2000, 300, 'Mean', fontsize=12)

# Median line
plt.axvline(median_salary, color='black', linestyle='dashed', linewidth=2)
plt.text(median_salary - 8000, 320, 'Median', fontsize=12)

# Labels & Title
plt.title("Salary Distribution", fontsize=16)
plt.xlabel("Salary")
plt.ylabel("Frequency")

# X-axis formatting
plt.xticks(np.arange(20000, 115000, 5000), rotation=45)

plt.tight_layout()
plt.show()
```
