```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[6,7,5,3,2]
plt.plot(x,y)
plt.show()
```

Title, Labels, Legends 

```python
x=[1,3,8,5,9] 
y=[2,5,2,6,8] 
plt.plot(x,y)
plt.title('bagdana')
plt.ylabel('tempal')
plt.xlabel('hotel')
plt.show()
```

Lengends

```python
x=[1,3,8,5,9] 
y=[2,5,2,6,8] 
 
x2=[2,3,5,7,6] 
y2=[3,6,2,6,8] 

plt.plot(x,y,label='line 1')
plt.plot(x2,y2,label='line 2')
plt.title('cdmi')
plt.ylabel('y axis')
plt.xlabel('x axis')
plt.legend()

plt.show()
```

color in graph

```python
import matplotlib.pyplot as plt 
x=[1,2,5,3,4,5,6] 
y=[2,4,6,8,9,11,13] 
x1=[3,5,7,9,1,5] 
y1=[3,4,5,6,8,2] 
plt.plot(x,y,label="line 1",color="green") 
plt.plot(x1,y1,label='line 2',color="black") 
plt.legend() 
plt.show() 
```

BAR CHARTS 

```python
x=[1,3,5,6,8]
y=[5,7,4,8,2]
plt.bar(x,y,label='line 1',color='black')
plt.title('cdmi')
plt.ylabel('y axis')
plt.xlabel('x axis')
plt.show()
```

HISTOGRAMS 

```python
year=[21,25,48,40,50]
man=[10,20,30,40,50,]
 
plt.hist(year ,man, histtype="bar",rwidth=0.7) 
 
plt.xlabel("X-Axis SIde") 
plt.ylabel("y-Axis SIde") 
 
plt.title("A Simple Histogram") 
plt.show() 
```

Scatter Plots / Scatter Graph 

```python
x=[1,3,5,6,8]
y=[5,7,4,8,2]
plt.bar(x,y,label='line 1',color='black')
plt.ylabel('y axis')
plt.xlabel('x axis')
plt.title("Scatter Plots / Scatter Graph") 
plt.legend() 
plt.show()
```

matplotlib.markers 

```python
x=[1,3,5,6,8]
y=[5,7,4,8,2]
plt.scatter(x,y,label='line 1',color='black',marker='x')
plt.ylabel('y axis')
plt.xlabel('x axis')
plt.title("Scatter Plots / Scatter Graph") 
plt.legend() 
plt.show()
```
